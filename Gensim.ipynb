```python
from gensim.models import Word2Vec
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Define the example sentences
sentences = [
    "Hi hanscey cronje nice to meet you, how you died, in the road accident or in the plane accident",
    "where is garry kristen"
]


# Split the sentences into lists of words
word_lists = [sentence.lower().split() for sentence in sentences]
word_lists
```

```python
# Train the Word2Vec model on the example sentences
model = Word2Vec(word_lists, min_count=1)

# Get the embeddings for "associate" and "employee"
embedding_1 = model.wv['hanscey']
embedding_2 = model.wv['garry']

# Print the embeddings
print("Embedding for 'hanscey': ", embedding_1)
print("Embedding for 'garry': ", embedding_2)
```

```python
embedding_1.shape
```

```python
similarity = cosine_similarity(embedding_1.reshape(1,-1), embedding_2.reshape(1,-1))[0][0]

print("Cosine similarity : ", similarity)
```


"word2vec-google-news-300" is a pre-trained word embedding model developed by Google which contains word vectors for more than 3 million words and phrases. The vectors in this model were trained on a large dataset of news articles and have 300 dimensions.

```python
import gensim.downloader as api

# Load a pre-trained Word2Vec model using the Gensim downloader
model = api.load("word2vec-google-news-300")
```

```python
dog = model['dog']
dog
```
